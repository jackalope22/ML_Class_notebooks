```python
products = products[products['rating']!= 3]
```

```python
products['sentiment'] = products['rating'] >= 4
```

```python
products
```

```python
train_data,test_data = products.random_split(.8,seed=0)
```

```python
selected_words = ["awesome_count", "great_count", "fantastic_count", "amazing_count", "love_count", "horrible_count", "bad_count", "terrible_count", "awful_count", "wow_count",
           "hate_count"]
```

```python
selected_word_model = tc.logistic_classifier.create(train_data,target='sentiment', features=selected_words, validation_set=test_data)
```

```python
selected_word_model.coefficients
```

```python
selected_word_model.evaluate(test_data)
```

```python
products['predicted_sentiment'] = selected_word_model.predict(products, output_type = 'probability')
```
